## SIMPLE SUMMARIES

We don’t always need complex tools.

Sometimes we just want to count things.

```python
cirights["Freedom of Speech"].value_counts()
```

WHAT IS THIS DOING?

.value_counts() counts how often each value appears.

So instead of averages, we now see:

* how many countries score 0
* how many score 1
* how many score 2

This is useful when variables have a small number of categories. We can also use value_counts to plot our data.

```python
cirights["Freedom of Speech"].value_counts().sort_values().plot(kind="bar")

plt.title("Freedom of Speech (2017)")
plt.show()
```

To get it in the right order we use the function sort_index() instead of sort_values, as sort index sorts by the value itself, instead of the counts.

We use .groupby() for this.

## RELATIONSHIPS IN THE DATA

So far we compared one thing at a time. Now let's look
at many variables at once.

### CORRELATION

Correlation tells us:

* which variables move together
* how strong that relationship is

**Important:**

* positive → they increase together
* negative → one goes up, the other goes down

This does NOT mean causation.

```python
corr = cirights[[
    "Physical Integrity Rights Index",
    "Freedom of Speech",
    "Judicial Independence",
    "Right to a Fair Trial",
    "Fair Trial Laws"
]].corr()

corr
```

## WHAT IS THIS DOING?

We select a few variables and calculate their correlation.

This gives us a table where:

each value shows how strongly two variables are related
values range from -1 to 1

Interpretation:

* close to 1 → strong positive relationship (they increase together)
* close to 0 → little or no relationship
* close to -1 → negative relationship (one goes up, the other goes down)

So now we can quickly see:

Which aspects of rights tend to go together across countries?

```python
## We can make a heatmap, to more easily see the relationship.


## Simple heatmap.

plt.imshow(corr, cmap="plasma")
plt.colorbar()

# Add axis labels
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
plt.yticks(range(len(corr.index)), corr.index)

plt.title("Correlation Matrix")
plt.show()
```

```python
# Fancy heatmap with numbers

fig, ax = plt.subplots(figsize=(6, 5))

# Draw the heatmap (colors)
heatmap = ax.imshow(corr, cmap="viridis")

# Add labels to axes
ax.set_xticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=45, ha="right")

ax.set_yticks(range(len(corr.index)))
ax.set_yticklabels(corr.index)

# Add numbers inside each cell
for row in range(len(corr.index)):
    for col in range(len(corr.columns)):
        value = corr.iloc[row, col]
        ax.text(col, row, f"{value:.2f}", ha="center", va="center", color="white")

# Title and color scale
plt.title("Correlation Matrix")
plt.colorbar(heatmap)

plt.show()
```

## WHAT IS GOING ON HERE?

We are doing three things:

1. Drawing a grid of colors (the heatmap)
2. Adding labels so we know what each row/column is
3. Writing the actual numbers inside each square

The only “new” thing here is the loop:

It goes through each cell and writes the correlation value on top of it.

You do NOT need to fully understand the loop.

Just know: it adds the numbers to the plot.
