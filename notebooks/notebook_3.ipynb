# Flow control
A flow control statement decides what to do based on whether its condition is True or False, and every flow control statement uses a condition. You’ll frequently write code that could be described in English as follows: “If this condition is true, do this thing, or else do this other thing.”

We can, with flow control, make a program, which doesn't just have conditions that don't do anything, but actually get it to do something! For example, our ticket inspector, might now do something if age < 18 or parental_approval == False.


## if

The most common type of flow control statement is the if statement. The code block after if will execute if the statement's condition is True, if it is False it is skipped.




```
age = 17
parental_approval = True

if age > 18 or parental_approval == True:
  print("Welcome")

if age < 18 and parental_approval == False:
  print(Sorry, you're not old enough to see this movie")
```
Now we are cooking! An if block can optionally be followed by an else statement, which only executes when the if statement's condition is False.



```
if age > 18 or parental_approval == True:
  print("Welcome")
else:
  print("You're not old enough, kiddo")
```
We can write much tighter code using the else keyword.

In the above code, only one condition is set, and if it is True then our code does something, and if it is false, it does something else. We can use another keyword **elif** which stands for "else if", where we get to make a condition that also must be true, for it to fire. It doesn't automatically fire, if the if statement is False (like the else statement does)


```
if name == "Tobias":
  print("Hi Tobias!)
elif name == "John"
  print("Hi John!)
else:
  print("Who're you?")
```

So here we have two conditions, we can make as many elif statements as we want. But as soon as a True condition has been met, the rest of the conditions are skipped over.

Using the while loop, we can make code execute again and again. It does not stop until the conditions that fired it is broken (... or we use a break statement, which you'll learn about shortly). To make it, we need the while keyword, a condition to be True, a colon, and a code block.

A while loop will continue to 're-run' while its condition is true. It can therefore be dangerous, and many programmers accidentally make infinite programs. To exit a program early press CTRL-C or Command-C for Mac people.

```
x = 1
while x == 1:
  print("Uh-oh")
```
This loop will run indefinitely-- or until you manage to stop it using control C, command C, or more drastic measures.

We can use a keyword to break out of loops early. These are called break statements, and if the execution reaches a break statement, it immediately exits the loop.

````python
counter = 0 #if I had put this into the loop, it would keep resetting counter to 0, and we would never leave it.

while True == True:
  print("Hi!")
  counter += 1 # this simply means add and set equal to.
  if counter > 5:
    break
````

## For Loops

The while loop keeps looping while its condition is True (which is the reason for its name), but what if you want to execute a block of code only a certain number of times? You can do this with a for loop statement.

The for loop takes the for keyword, a variable name, the in keyword, and a call to a 'range'.

We saw this code last time:
```
words_from_text = text.lower().split()

clean_words = []

for word in words_from_text:
  if word not in stop_words:
    clean_words.append(word)
```
In this case, 'word' is the variable we use for each 'part' of words_from_text, and words_from_text is our range as well. If there are 5 words in the variable, it will run 5 times.

Often, we don't have a specific list with a number we want to run, we may also only want the first 30 words. We therefore sometimes use the range() function.

the range function gives us basically a sequence of numbers. We can tell it where to start (it will include it), and must tell it where to stop (it will not include that).


````python
print('Hello!')
for i in range(5): #
    print('On this iteration, i is set to ' + str(i))
print('Goodbye!')
````

Python starts counting from 0. So if we want it to print 1 2 3 we must say



```
for n in range(1,4):
  print(n)
```



There is one more break statement, which instead of getting us out of the loop 'resets' it; think of it as more that it completes it, but the code beneath it is then not run (for that iteration). That is the keyword 'continue'

````python
unfortunate_numbers = [4, 9, 13]

for n in range(1,11):
  if n in unfortunate_numbers:
    continue
  print(n)

#if a number we don't want printed is the same as n, then the if statement condition is true and we reach the 'continue' and resets us
#the print(n) thus doesn't fire, and we skip it.
````

````python
words_of_speech_trump = trump_speech.lower().split() #string methods to make all words lower_case and put each wordi nto a list
words_of_speech_obama = obama_speech.lower().split()




## Functions that we will use

def average_word_length(speech): #make a function that calculates average word length, because we use it multiple times
  container_for_word_length = 0 # keep this out of the for loop, otherwise it will keep resetting to 0 in the loop
  for word in speech:
    container_for_word_length += len(word) #for each word, we calculate the length
  average_words = container_for_word_length / len(speech) #simple average divided by
  print(average_words)

def delete_word(speech, word_to_delete, potential_other_word=None, potential_third_word=None):
  new_list = []
  for word in speech:
    if word != word_to_delete and word != potential_other_word and word != potential_third_word:
      new_list.append(word)
  return new_list

##Average word length before we work our magic

#average word_length for trump
print("this is the average for Donald (before the magic): ")
average_word_length(words_of_speech_trump)

#average word_length for obama
print("this is the average for Obama (before the magic):")
average_word_length(words_of_speech_obama)


##Our data isn't perfectly clean though. Obama's speech has a lot of "--" and Trump has "-".
#We also have "(applause.), and "(applause).", and "(laughter)" ->
# We could probably write some code that deletes words that start with ( or ends with them, but let's just use our delete_word function
clean_speech_obama = delete_word(words_of_speech_obama, "(applause.)", "(applause)", "--")

if "--" or "(applause)" or "(applause.)" not in clean_speech_obama:
  print("We've succesfully removed the filler words from Obama's speech.")


clean_speech_trump = delete_word(words_of_speech_trump, "(applause.)","-", "(laughter.)")

if "-" or "(applause.)" or "(laughter.)" not in clean_speech_trump:
  print("We've succesfully removed the filler words from Trump's speech.")

## Let's recompare average word length after having cleaned the data

print("This is the real average word length for Trump:")
average_word_length(clean_speech_trump)

print("This is the real average word length for Obama:")
average_word_length(clean_speech_obama)


print()
````
