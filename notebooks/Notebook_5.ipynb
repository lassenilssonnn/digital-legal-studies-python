## .loc → Label-based selection

.loc selects rows and columns using their names or labels. This is more like selecting values from a dictionary.

Example:



```
data.loc[0] #→ row with index label 0
```

```
data.loc[0:5, ["Country", "year"]] #→ rows 0–5, only the columns Country and year
```

Important rule: **.loc includes the final value.**



```
data.loc[0:5]
```
returns rows:

```

```

**.Iloc** precludes the final value (like normal Python) so 0:5, is 0, 1, 2, 3, 4.

## Using Conditions to Filter Data

Now we combine two ideas we just learned:

1. Selecting data with `.loc`
2. Creating **conditions**

Look at this expression:

```
data.Country == "Turkey"
```

This checks **every row** in the `Country` column and asks:

> "Is this row Turkey?"

Instead of returning a single value, pandas returns a **Series of True/False values**, one for each row.


| Country | Result |
| ------- | ------ |
| Turkey  | True   |
| Turkey  | True   |
| France  | False  |
| Germany | False  |

So pandas creates something like:

```
True
True
False
False
False
...
```

This is called a **boolean mask**.


## Why this is powerful

We can use this mask to **filter the dataframe**.
'
Example:

```
data.loc[data.Country == "Turkey"]
```

This means:

> "Give me the rows where the condition is True."

So pandas keeps only rows where the country is **Turkey**.


## Combining conditions

We can also combine conditions.

Example:

```
data.loc[(data.Country == "Turkey") & (data.Judgement_type == "violation")]
```

This means:

* Country must be **Turkey**
* AND the judgement type must be **violation**

The symbol `&` means **AND**.

This lets us ask very specific questions of the data.

---

## Why this matters

Filtering data like this is one of the **most important skills in data science**, because it lets us answer questions like:

* How many cases involve **Turkey**?
* Which **countries had violations after 2020**?
* Which cases involve **Article 10**?

We are no longer just looking at the data — we are **querying it**.


````python
data.Country == "Turkey" ## gives us true and false with indexes!!
````

````python
data.loc[(data.Country == "Turkey") & (data.Judgement_type == "violation")] # use ambersand to bring them together
````

````python
# here we also pick which columns we want to see and have 2 conditions
data.loc[(data.Country == "Denmark") & (data.year > 2020), ["Country", "year", "Judgement_type", "Definition"]] #
````

````python
## here we use | which means or so EITHER of the conditions has to be true

data.loc[(data.Country == "Denmark") | (data.year == 1989),["Country", "Importance", "year"]] # or, sorry, pandas uses annoying symbols
````

## `.isin()` and `.isnull()`

Sometimes we want to check **multiple values** or find **missing data** in our dataframe.

Pandas gives us useful tools for this.

---

## `.isin()` → check if a value is in a list

Instead of writing multiple conditions like this:

```
data.Country == "France" OR data.Country == "Italy"
```

we can use `.isin()`.

Example:

```
data.Country.isin(["France", "Italy"])
```

This checks every row in the `Country` column and asks:

> "Is this value France or Italy?"

The result is again a **True/False series**.

Example idea:

| Country | Result |
| ------- | ------ |
| France  | True   |
| Italy   | True   |
| Spain   | False  |

We can then use it to filter the dataframe:

```
data.loc[data.Country.isin(["France", "Italy"])]
```

This returns **only rows where the country is France or Italy**.

---

## `.isnull()` → find missing values

Real datasets often contain **missing information**.

In pandas, missing values are called **NaN** (Not a Number).

To detect them we use:

```
data.Importance.isnull()
```

This again returns **True/False** values:

| Importance | Result |
| ---------- | ------ |
| 3          | False  |
| 2          | False  |
| NaN        | True   |

---

## What can we do with this?

We can **filter rows where data is missing**.

Example:

```
data.loc[data.Importance.isnull()]
```

This shows **all rows where the Importance value is missing**.

We can also inspect specific columns:

```
data.loc[data.Importance.isnull(), ["Country", "Importance"]]
```

This helps us **identify problems in the dataset** and decide what to do next.

Handling missing data is an important part of real-world data analysis.


````python
## Using is in and filtering two columns


data.loc[data.Country.isin(["France", "Italy"]), ["Country", "Importance"]]
````

````python
# Doing this, we see something weird. key_cases? Let's remove that.
data = data.loc[data.Importance != "key_cases"]
````

````python

data.loc[data.Importance.isnull(), ["Country", "Importance"]]
````

## What should we do with missing values?

When we find `NaN` values in a dataset, it means **some information is missing**.

This is extremely common in real-world data.

There is no single correct solution — analysts usually choose between three common approaches.

---

## 1. Inspect the missing data

First, we often **look at where the missing values are**.

Example:

```
data.loc[data.Importance.isnull(), ["Country", "Importance"]]
```

This helps us understand **which observations are missing data**.

---

## 2. Remove rows with missing values

If the missing information is not important, we may simply **remove those rows**.

Example:

```
data.dropna(subset=["Importance"])
```

This removes rows where the `Importance` value is missing.


## 3. Replace missing values

Sometimes we want to **fill the missing values with something else**.

Example:

```
data["Importance"] = data["Importance"].fillna(0)
```

This replaces missing values with `0`.


## Which method is best?

It depends on the **research question** and the **importance of the variable**.

In empirical research, handling missing data carefully is an important part of the analysis process.

## Using `.map()` to Transform Data

Sometimes we want to **apply a function to every value in a column**.

We can do this with `.map()`.

Example:

```
data.Document_Title.map(lambda x: x.lower())
```

This does the following:

1. Takes every value in `Document_Title`
2. Applies a function to it
3. Returns the transformed values

---

## What is `lambda`?

A **lambda function** is a small anonymous function.

Example:

```
lambda x: x * 10
```

This means:

> "Take `x` and multiply it by 10."

So when we write:

```
data.Document_Title.map(lambda x: x.lower())
```

we are saying:

> "Take every document title and convert it to lowercase."

---

## Why this matters

Transforming columns lets us **prepare data for analysis**.

For example we might:

* standardize text
* create new legal categories
* generate variables used in statistical analysis

In data science, this step is often called **data preprocessing**.


````python
# Exercise
# Create a new column called "Country_lower"
# that contains the Country column in lowercase

data["Country_lower"] = data.Country.map(lambda x: x.lower())
````

## Grouping Data with `groupby()`

So far we have summarized **single columns**.

For example:

```python
data["Country"].value_counts()
```

This tells us how many cases each country has.

But often we want to ask **more complex questions**, such as:

* How many cases occur **per year**?
* What is the **latest case year for each country**?
* How many cases does each **country have per year**?

To answer questions like this, we use **`groupby()`**.

---

## Basic idea

`groupby()` splits the dataset into **groups that share the same value**.

Another method tells pandas what to do with each group.


Example:


````python
data.groupby("year").year.count()
````


This means:

1. Group the dataset by **year**
2. Count how many rows belong to each year

---

## Another example


````python
data.groupby("Country").year.min()

# You can think of each group we generate as being a slice of our DataFrame containing only data with values that match.
````


This means:

1. Group rows by **Country**
2. Look at the **year column**
3. Find the **minimum year** for each country

This tells us the **oldest case year** for each country.

---

## Why this matters

Grouping lets us **summarize data across categories**, which is one of the most important tools in data analysis.

Instead of looking at individual cases, we can now understand **patterns across the dataset**.

## `.replace()` → fixing or standardizing values

Sometimes datasets contain **codes or abbreviations** instead of readable labels.

For example, the value `"HEJUD"` is a **code used in the dataset** for a type of document.

To make the data easier to understand, we can replace this code with a clearer label.

Example:


Now the dataset itself is updated.

Cleaning and standardizing values is a common step in data analysis.

## Renaming columns

Sometimes column names in a dataset are **unclear, inconsistent, or hard to read**.
Pandas lets us rename them using `.rename()`.

Example:

```python
data.rename(columns={'date': "Date", "year":"Year", "label":"Label", "all_labels":"All_labels"})
```

This means:

* `"date"` becomes **`"Date"`**
* `"year"` becomes **`"Year"`**
* `"label"` becomes **`"Label"`**
* `"all_labels"` becomes **`"All_labels"`**

We pass a **dictionary** where:

* the **key** is the current column name
* the **value** is the new column name.

### Important detail

Like many pandas operations, `.rename()` **does not change the dataframe unless we assign it back**.

So the usual pattern is:

```python
data = data.rename(columns={'date': "Date", "year":"Year", "label":"Label", "all_labels":"All_labels"})
```

This updates the dataframe with the new column names.


````python
data.rename(columns={'date': "Date", "year":"Year", "label":"Label", "all_labels":"All_labels"})
````
