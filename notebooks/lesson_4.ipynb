## Flow Control

Programs often need to make decisions.
In Python, we control which code runs by using conditional statements.

An if statement evaluates a condition.
If the condition is True, the code inside the block runs.

We can add additional conditions using elif (short for else if).
These are only checked if the previous conditions were not satisfied.

Finally, we can use else.
The else block runs if none of the previous conditions are true.

```
def get_food():
  #insert code in indented block that makes us go make food from the fridge

def shop_for_food():
  #insert code in indented block that makes us shop for food

hungry = True

if hungry and food_in_fridge:
  get_food()

elif food_in_fridge == False:
  shop_for_food()

else:
  print("I'm happy I have food in my fridge and that I'm not hungry")

```
Python checks the conditions from top to bottom.

Is hungry and food_in_fridge true?

Only one branch of an if / elif / else chain will execute.

## Indexing:
Accessing Data by Position

In previous lessons, we used lists. For example:

cases = [
    "Marbury v. Madison",
    "Brown v. Board of Education",
    "Roe v. Wade",
    "Obergefell v. Hodges"

A list is an ordered collection. That means: Every element has a position. That position is called its index.

In Python, indexing starts at 0, not 1. Indexes start at 0 because they represent distance from the start.

Index 0 means:

Zero steps away from the beginning.

Index 3 means:

Three steps away from the beginning.

````python
cases = [ "Marbury v. Madison", "Brown v. Board of Education", "Roe v. Wade", "Obergefell v. Hodges" ]

# So:

print(cases[0])  # First case
print(cases[1])  # Second case
print(cases[3])  # Fourth case
````

## Slices

Just as an index can get a single value from a list, a slice can get several values from a list, in the form of a new list. We enter a slice between square brackets, like an index, but include two integers separated by a colon. Notice the difference between indexes and slices:

spam[2] is a list with an index (one integer).

spam[1:4] is a list with a slice (two integers).

In a slice, the first integer is the index where the slice starts.

The second integer is the index where the slice ends. The list created from a slice will go up to, but will not include, the value at the second index. This is the same with the len(2, 5). It will start from 2 and go up to 4. It will not include 5.

For example,

```
numbers = [0, 1, 2, 3, 4, 5, 6]

print(numbers[1:3])

```
this will give us the numbers 1, 2.

# COMPUTATIONAL FRICKIN' SCIENCE.


Now we're going to do actual computational legal science.

I have downloaded a dataset called "final_for_viz.csv". It is a CSV file, meaning it uses comma seperated values

A CSV file is just a plain text file where each row is a line and each value is separated by a comma.

example:


```
Document_Title,Country,Year,Article
CASE OF EREN v. TURKEY,Turkey,2003,6
CASE OF HANDYSIDE v. UNITED KINGDOM,United Kingdom,1976,10
```

The first line is the header row.
These are the column names.

When we use csv.DictReader, Python reads this so that each row becomes a dictionary where the headers become the keys.



````python
import csv

with open("final_for_viz.csv", newline="") as file:
    reader = csv.DictReader(file)
    data = list(reader)
````

When we run several things happen: We import the csv module. This module contains tools for working with CSV files.

open("final_for_viz.csv") opens the file so Python can read it.

csv.DictReader(file) reads the CSV file and converts each row into a dictionary.

Each column name becomes a key, and the value in that column becomes the value.

list(reader) stores all rows in a list called data.

This means that data is now a list of dictionaries.

Each dictionary represents one case.

So conceptually we now have something like this:

[
{case_1_data},
{case_2_data},
{case_3_data},
]

This structure is extremely common in data science and computational social science.
It allows us to loop through cases and access information using dictionary keys such as "Country", "year", "Article", and "Document_Title".

## Which Country Has the Most Article 6 Violations?

Article 6 of the European Convention on Human Rights protects the right to a fair trial.

Let's task is to investigate the dataset and answer the following question:

````python
article6_country_counts = {}

for row in data:

    if row["Article"] == "6" and row["Judgement_type"] == "violation":

        country = row["Country"]

        if country not in article6_country_counts:
            article6_country_counts[country] = 1
        else:
            article6_country_counts[country] += 1

print(article6_country_counts)


max_country = max(article6_country_counts, key=article6_country_counts.get)

print("Country with most Article 6 violations:", max_country)
print("Number of violations:", article6_country_counts[max_country])
````

````python
most_common_article = max(article_counts, key=article_counts.get)

print("Most common article:", most_common_article)
print("Number of cases:", article_counts[most_common_article])
````


The dataset includes the year of each judgment.

````python
year_counts = {}

for row in data:

    year = row["year"]

    if year not in year_counts:
        year_counts[year] = 1
    else:
        year_counts[year] += 1

year_counts
````

````python
most_cases_year = max(year_counts, key=year_counts.get)

print("Year with most cases:", most_cases_year)
print("Number of cases:", year_counts[most_cases_year])
````

So far, we have used Python to:

* load a dataset

* inspect structured data

* loop through observations

* count legal variables

But researchers rarely stop at raw numbers. We often want to visualize patterns in data. Let’s create a simple graph showing:

How many ECtHR cases appear each year in this dataset.

````python
# Step 1: Count cases per year

year_counts = {}

for case in data:
    year = int(case["year"])

    if year not in year_counts:
        year_counts[year] = 1
    else:
        year_counts[year] += 1

#This creates a dictionary like:
#{1995: 3, 1996: 7, 1997: 12, ...}
````

````python
# 2: Convert the Dictionary into Sorted Lists

years = sorted(year_counts.keys()) #returns all the years in a list
counts = [] # start with an empty list

for y in years: # go through each year in the years list
  value = year_counts[y] # look up how many cases happened that year
  counts.append(value) # add that number to the counts list

print(years)
print(counts)
````
